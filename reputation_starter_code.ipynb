```python
def old_paper(interactions, phi):
    rj = 0
    sj = 0
    for i in interaction:
        
        rj += i[0]
        sj += i[0]

    rj += 1
    rj *= phi
    sj -= 1
    sj *+ phi
    
    return rj, sj
```

```python
def new_paper(interactions, phi):
    rj = 0
    sj = 0
    for i in interaction:
        
        rj += (1+ i[0]) * phi
        sj += (1- i[0]) * phi 
        
    return rj, sj
```

```python
def new_paper_with_decay(interactions, phi, decay):
    rj = 0
    sj = 0
    exp_val_n = len(interactions)
    count = 0
    for i in interactions:
        count += 1
#         print(i)
        rj += ((1+ i) * phi) * (decay **(exp_val_n - count)) 
        sj += ((1- i) * phi) * (decay **(exp_val_n - count))
        print(', st ='+ str(rj)+', ds = '+str(sj))
        
    return rj, sj
```

```python
##get common vendors between requester and adviso
def find_vedors(df, requestor):
    #get requestors neighbors == vendors [xx,xx,xx, xx]
    #for all neighbors get their vendors [yy, yyy, yy, yy, yy]
    #get intersection of neigbors with vendors
    req_list1 = df[df['userID1'] == requestor]
    req_list2 = df[df['userID2'] == requestor]
    
#     print(df.loc[(df['userID1'] == requestor ) | (df['userID2'] == requestor )].userID2.unique())
#     print(df.loc[(df['userID1'] == requestor ) | (df['userID2'] == requestor )].userID1.unique())
    
        
    print('neighbors in req_list1 are '+str(req_list1.userID2.unique()))
    print('neighbors in req_list2 are '+str(req_list2.userID1.unique()))
    all_neighbors = np.concatenate((req_list1.userID2.unique(), req_list2.userID1.unique()))
    print('all_neighbors '+ str(all_neighbors))
    
    neighbor_list = {}
    for n in all_neighbors:
        print('neighbor '+str(n)+'\'s vendors ')
#         print(df[df['userID1'] == n ].userID2.unique())
#         print(df[df['userID2'] == n ].userID1.unique())
        n_vendors = np.concatenate((df[df['userID1'] == n ].userID2.unique(), df[df['userID2'] == n ].userID1.unique()))
        n_vendors = list(filter(lambda a: a != requestor, n_vendors))
        neighbor_list[n] =  n_vendors
        print('n_vendors '+ str(n_vendors))
    
    ##for  k neigbor get common vendor
    print(neighbor_list)
#     print(len(neighbor_list))
    comm_vendors = {}
    for k, ven in neighbor_list.items():
#         print('kkkkkkkkk')
#         print(k)
#         if k == 102.0:
#             del ven[-1]
        comm_vendors[k] =  [i for i, j in zip(all_neighbors, ven) if i == j]
        
    print('comm_vendors')
    print(comm_vendors)
    
    return comm_vendors
          
#     for idx, row in neighbor:
#     print(neighbors.userID2.unique())

requestor = 101
print('requestor '+ str(requestor))
find_vedors(df_dbu, requestor)
```

```python
def honesty(df, user=2):
    honDF = pd.DataFrame()
    for k, af in df:
        print(af)
        hon_li = []
        cum_cumsat = 0
        cum_cumdis = 0
        for index, row in af.iterrows():
            hon = 0
            cum_cumsat += row['U'+str(user)+ 'cum_sat']
            cum_cumdis += row['U'+str(user)+ 'cum_dis']
            hon = cum_cumsat / (cum_cumsat + cum_cumdis)
            print('hon')
            print(hon)
            hon_li.append(hon)

        af['honesty'+str(user)] = hon_li
        print(af)
        honDF = honDF.append(af, ignore_index = True)
    return honDF
```

```python
def init_trust(df, user=2, beta=0.5):
    iniTDF = pd.DataFrame()
    for k, af in df:
        print(af)
        ini_trust_li = []
        cum_repu = 0
        trust_propensity = random.uniform(0, 1)
        tp_li = []
        for index, row in af.iterrows():
            ini_trust = 0
            cum_repu = row['repu'+str(user)]
            ini_trust =  (beta * trust_propensity ) + ((1-beta) * cum_repu )
            print('ini_trust')
            print(ini_trust)
            ini_trust_li.append(ini_trust)
            tp_li.append(trust_propensity)

        af['init_T'+str(user)] = ini_trust_li
        af['tp'+str(user)] = tp_li
        print(af)
        iniTDF = iniTDF.append(af, ignore_index = True)   
    return iniTDF
```

```python
def trust_tDash(df, user=2, beta=0.5):
    tDtDF = pd.DataFrame()
    for k, af in df:
        print(af)
        trust_li = []
        trust_dash_li = []
        sum_test = []
        
        trust_propensity = random.uniform(0, 1)
        for index, row in af.iterrows():
            trust = 0
            trust_dash = 0
            sum_t = 0
            t = row['init_T'+str(user)]
            b = row['bel'+str(user)]
            d = row['dis'+str(user)]
            u = row['unc'+str(user)]
            trust = b + (t * u)
            trust_dash = d + ((1 - t) * u) 
            sum_t = trust + trust_dash
            print('sum_t')
            print(sum_t)
            trust_li.append(trust)
            trust_dash_li.append(trust_dash)
            sum_test.append(sum_t)

        af['T'+str(user)] = trust_li
        af['T_dash'+str(user)] = trust_dash_li
        af['sum_tdt'+str(user)] = sum_test
        print(af)
        tDtDF = tDtDF.append(af, ignore_index = True)   
    return tDtDF
```
