```python

def get_uid(vd, phi, decay, user):
    newDF = pd.DataFrame()
        
    for k, uid in vd:
        itr_list = uid[user+'rating'].tolist() 
#         print(itr_list)
        cum_rj = 0
        cum_sj = 0
        exp_val_n = len(itr_list)##TODO len(itr_list - forgetnumber)
        count = 0
        sj_list = []
        rj_list = []
        cum_sj_list = []
        cum_rj_list = []
        
#         for i in itr_list:
        print('=========================================================')
        for index, row in uid.iterrows():
            count += 1
            rj = ((1- row[user+'rating']) * phi) * (decay **(exp_val_n - count)) #disatisfaction
            sj = ((1+ row[user+'rating']) * phi) * (decay **(exp_val_n - count)) #satisfaction
            cum_rj += rj
            cum_sj += sj
            

#             print('rj ='+ str(rj)+', sj = '+str(sj))
            sj_list.append(sj)
            rj_list.append(rj)
            cum_sj_list.append(cum_sj)
            cum_rj_list.append(cum_rj)
            
            
        
#         print('vendor = '+ str(k)+', st ='+ str(rj)+', ds = '+str(sj))
        uid[user+'dis'] = rj_list  #disatisfaction list
        uid[user+'sat'] = sj_list #satisfaction list
        uid[user+'cum_dis'] = cum_rj_list #cummilative disatisfaction list
        uid[user+'cum_sat'] = cum_sj_list #cummilative satisfaction list
        

        print(uid)
        newDF = newDF.append(uid, ignore_index = True)
    return newDF

```

```python
#old paper  anaka function and bayes function
def anaka_bayes(df, anaka= True,  user='Consumer'):
    theta = random.uniform(0, 1)
    initr = (1 + theta)/2
    inits = (1 - theta)/2
    
    honDF = pd.DataFrame()
    df.reset_index(drop=True, inplace=True)
    cd = df.groupby(['Consumer', 'Provider'])
    if (not anaka):
        cd = df.groupby(['Consumer'])
        initr = 1
        inits = 1
    
    for k, af in cd:
        print(af)
        hon_li = []
        cum_cumsat = 0
        cum_cumdis = 0
        countrij = 1
        for index, row in af.iterrows():
            expectedtrust = 0      
            if countrij == 1: 
                cum_cumsat += initr
                cum_cumdis += inits
             
            countrij += 1
            cum_cumsat += row['Consumer'+ 'cum_sat']
            cum_cumdis += row['Consumer'+ 'cum_dis']
#             hon = cum_cumsat / (cum_cumsat + cum_cumdis)
            expectedtrust =  (cum_cumsat) / ( cum_cumsat + cum_cumdis)
            print('expectedtrust')
            print(expectedtrust)
            hon_li.append(expectedtrust)

        
        if (anaka):
            af['anakaE'] = hon_li
        else:
            af['bayesT'] = hon_li            
        
        print(af)
        honDF = honDF.append(af, ignore_index = True)
    return honDF
```

```python
def credibility_oldpaper_n_trust(df):
    credDF = pd.DataFrame()
    cd = df.groupby(['Consumer'])
    for k, af in cd:
#         provider_count = af.Provider.unique().count()
        provider_count = 1
        provider = 1
        negatives = 0
        totallenght = 0
#         credibility = 0
        credlist = []
        trustT = []
#         totalpostives = af.ProviderPostives.sum()
#         totallength = af.length()
        for index, row in af.iterrows():
            if (provider != row.Provider and  provider_count != 1):
                provider_count += 1
            provider = row.Provider
            
            print(provider_count)
                
            totallenght += 1
            if row.ProviderPostives < 1:
                negatives +=1
            credibility = 1 - (negatives/totallenght)
            credlist.append(credibility)
            trust = credibility * row.anakaE
            trustT.append(trust)
            
        af['Consumercred'] = credlist
        af['anakaTrust'] = trustT
        credDF = credDF.append(af, ignore_index = True)
    return credDF
        
```

```python
def reputation3(df, sim, id_idx, idx_id, commonv, comv_advisors, user='Consumer',threshold = 0.5):
    uniqueid1list = df.Consumer.unique().tolist()
    uniqueproducersl = df.Provider.unique().tolist()
    rep_matrix = [[0 for i in range(len(uniqueid1list))] for j in range(len(uniqueproducersl))]
    print(commonv)
    
    mf = df.groupby(['Consumer', 'Provider']) 
    pf = df.groupby(['Provider'])
    
    minli = [random.randrange(2, len(uniqueid1list)) for i in range(len(uniqueid1list))]
    
    vidx_vid = {}
    vid_vidx = {}
    count = 0
    for i in uniqueproducersl:
        vidx_vid[count] = i
        vid_vidx[i] = count #mostly used
        count += 1
    print(vidx_vid)
    print(vid_vidx)
    
    v_neigbors = {}
    for y, gf in pf:
        print(gf.Consumer.unique())
        v_neigbors[y] = gf.Consumer.unique().tolist()
    

    print('len(v_neigbors)'+str(len(v_neigbors)))
    print('len(uniqueproducersl'+str(len(uniqueproducersl)))
    
    

    vra_rep = {} #(vendor requestor advisor) = reputaion
    for vendor in v_neigbors:
#         print(key)
        reqindex = 0
        for requestor in uniqueid1list:
#             print(v_neigbors[vendor])
            hon = 0
            rep = 0
            hon_count = 0 #this is m in equation
            rep_count = 0 #this in n in equation
            ksim = sum(1 for i in sim[id_idx[requestor]] if i > threshold)
            print('sum(1 for i in sim[id_idx[requestor]] if i > threshold)')
            print(ksim)
            minimumadvisors = minli[reqindex]
            reqindex += 1
            if ksim < minimumadvisors:
                alpha = ksim/minimumadvisors
            else:
                alpha = 1

            for advisor in v_neigbors[vendor]:

                if((advisor,requestor) in commonv):

                    ab_sim = sim[id_idx[requestor]][id_idx[advisor]]
                    cum_result = 0
                    print('...commvl'+str((advisor,requestor)))
                    for x, af in mf:
                        if(x[0]==advisor and x[1]==vendor):
                            av_bel = af.Consumerbel.tail(1).item()
                            if ab_sim > threshold:
                                rep += (ab_sim * av_bel)
                                rep_count += 1

                else:
                    print('no common pairs '+ str((advisor,requestor)))
                    print('so we want '+ str(vendor)+' honesty with '+str(advisor))
                
                    for y, jf in mf: #in this loop following condition will occur only once
                        if(y[0]==advisor and y[1]==vendor): 
                            av_bel = jf.Consumerbel.tail(1).item()
                            va_hon = jf.honesty.tail(1).item()
                            hon += (av_bel * va_hon)
                            hon_count += 1
            reputation = 0
            if rep_count > 0:
                reputation = (alpha * rep)/rep_count
            
            if hon_count > 0:
                reputation += ((1 - alpha) * hon)/hon_count
            
            vra_rep[(vendor, requestor)] = reputation
    return vra_rep                    
            
```

```python
def init_trust(df, rep, beta=0.5):
    requestor_li = df.Consumer.unique().tolist()
    vendor_lis = df.Provider.unique().tolist()
    rv_initT = {}
    for vendor in vendor_lis:
        for requestor in requestor_li:
            trust_propensity = random.uniform(0, 1)
            r = 0
            if (vendor, requestor) in rep:
                r = rep[(vendor, requestor)]
            initT = (beta * trust_propensity ) + ((1-beta) * r)
            rv_initT[(requestor,vendor)] = initT
    return rv_initT
```
